```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Configura la ruta de la imagen
ruta_imagen = "../../../Imgs/Sento.jpg"  # ¡Cambia esto!

# Carga y muestra la imagen
img = mpimg.imread(ruta_imagen)
plt.figure(figsize=(8, 8))  # Tamaño de la figura
plt.imshow(img)
plt.axis('off')  # Oculta ejes
plt.title("Mi Crush ❤️")
plt.show()
```
